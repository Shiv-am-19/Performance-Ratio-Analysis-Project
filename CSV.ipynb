```python
import os
import pandas as pd
```

```python
pr_folder = 'C:/SHIV/Major Project/Finall Projects/Internship/PR'
ghi_folder = 'C:/SHIV/Major Project/Finall Projects/Internship/GHI'
```

```python
all_dataframes = []
```

```python
def read_csvs_from_folder(folder_path):
    for file in os.listdir(folder_path):
        if file.endswith('.csv'):
            file_path = os.path.join(folder_path, file)
            df = pd.read_csv(file_path)
            df['SourceFolder'] = os.path.basename(folder_path)
            all_dataframes.append(df)
            print(f"Loaded: {file}")
```

```python
read_csvs_from_folder(pr_folder)
read_csvs_from_folder(ghi_folder)
```

```python
combined_df = pd.concat(all_dataframes, ignore_index=True)
combined_df.to_csv('combined_output.csv', index=False)
print("Combined CSV saved as 'combined_output.csv'")
```
